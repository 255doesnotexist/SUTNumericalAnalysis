```python
def lagrange_interpolation(x, points):
    n = len(points)
    result = 0.0
    for i in range(n):
        term = points[i][1]
        for j in range(n):
            if i != j:
                term *= (x - points[j][0]) / (points[i][0] - points[j][0])
        result += term
    return result

points = [(1, 3), (2, 5), (3, 7)]

x = 4

interpolated_value = lagrange_interpolation(x, points)
print("Interpolated value at x =", x, ":", interpolated_value)
```
